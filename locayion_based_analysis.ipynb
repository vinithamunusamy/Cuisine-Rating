```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv("/content/sample_data/Dataset .csv")
print(df.head())
print(df.columns)
```

```python
plt.figure()
plt.scatter(df['Longitude'], df['Latitude'])
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Geographical Distribution of Restaurants")
plt.show()
```

```python
city_counts = df['City'].value_counts()

print("Top 10 Cities with Highest Number of Restaurants:")
print(city_counts.head(10))
```

```python
city_stats = df.groupby('City').agg(
    Average_Rating=('Aggregate rating', 'mean'),
    Average_Price_Range=('Price range', 'mean'),
    Total_Restaurants=('Restaurant ID', 'count')
)

# Sort by restaurant count
city_stats = city_stats.sort_values(by='Total_Restaurants', ascending=False)

print(city_stats.head(10))
```

```python
city_cuisine = df.groupby('City')['Cuisines'].apply(
    lambda x: x.value_counts().head(3)
)

print("Top Cuisines by City:")
print(city_cuisine.head(20))
```

```python
top_cities = city_counts.head(10)

plt.figure()
top_cities.plot(kind='bar')
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
plt.title("Top 10 Cities with Most Restaurants")
plt.show()
```

```python
print("INSIGHTS:")
print("- Restaurants are highly concentrated in metropolitan cities.")
print("- New Delhi, Gurgaon, and Noida dominate the dataset.")
print("- Cities with more restaurants generally maintain good average ratings.")
print("- Urban regions show more cuisine diversity and pricing flexibility.")
```
